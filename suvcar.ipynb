# Train Data

```python
#defining our features and target variable
x=suv.iloc[:,[2,3]].values #all row nd 2nd, 3rd column ,iloc used for integer based indexing
y=suv.iloc[:,4].values #all row nd 4 column
```

```python
#now checking my feature variable
x
```

```python
#now checking my target variable
y
```

```python
# Splitting the dataset into the Training set and Test set - 75-25 split
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0) ##random state ensure the same result,means same sample taken whenever run the code

```

```python
#whenever you working with large dataset then it's affecting in scaling or fitting so,we scale them in train nd test
#range of estimated salary and age is different so we scale the dataset
##Feature scaling as range of estimated salary and age is different
from sklearn.preprocessing import StandardScaler #preprcessing is a method to transform ur data
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0) #classifying
classifier.fit(x_train,y_train)
```

```python
#predicting text results
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

just add 65+24 and divide by 65+24+8+3 and we get the accuracy score

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
